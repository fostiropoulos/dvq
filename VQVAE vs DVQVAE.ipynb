```python
import os
os.environ['CUDA_VISIBLE_DEVICES']='1'
import tensorflow as tf
tf.keras.backend.clear_session() 
from keras.datasets import cifar10
(X,_),(_,_)=cifar10.load_data()
```

```python
from dvq_model import VQVAE
optimizer = tf.keras.optimizers.Adam(learning_rate=2e-4)
vqvae=VQVAE(D=255,K=512,L=10,mse=False)
vqvae.compile(optimizer=optimizer,loss=vqvae.losses)
vqvae.fit(X,epochs=100,batch_size=128)
```

```python
from dvq_model import DVQVAE
optimizer = tf.keras.optimizers.Adam(learning_rate=2e-4)
dvqvae=DVQVAE(D=255,K=512,L=10,mse=False)
dvqvae.compile(optimizer=optimizer,loss=dvqvae.losses)
dvqvae.fit(X,epochs=100,batch_size=128)
```
